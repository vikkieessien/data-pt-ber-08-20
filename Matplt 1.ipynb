```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

plt.plot(x, y, '')
plt.title("This is a plot.")

plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

### student portal example ###

```python
x1 = np.arange(0,10,1)
x2 = np.linspace(0,5,10)
plt.plot(x1, x1, label='linear')
plt.plot(x1, x2**2, label='quadratic')

plt.xlabel('x label')
plt.ylabel('y label')

plt.title("Linear vs Non Linear Plots")
plt.legend()
plt.show()
```

### plotting Multiple Series ###

```python
z = [5, 4, 3, 2, 1]

plt.plot(x, y, color = "magenta")
plt.plot(x, z, color = "yellow", linewidth = 5)
plt.legend(['y', 'z'])
plt.show()
```

### Scatter Plots ###

```python
a = np.random.random(10)
b = np.random.random(10)

plt.scatter(a,b)
plt.xlabel("A")
plt.ylabel("B")
plt.show()
```

### Pie Chart ###

```python
animals = ['Cat', 'Dog', 'Hamster']
values = [20, 30, 50]

plt.pie(values, labels = animals, colors = ['yellow', 'blue', 'green'])
plt.show()
```

### Bar Chart ###

```python
plt.bar(animals, height = values)
plt.show()
```

```python
# sideways chart, it changes to width, not height

plt.barh(animals, width = values)
plt.show()
```
