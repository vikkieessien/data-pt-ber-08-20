```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vehicles = pd.read_csv('vehicles.csv')
vehicles.head()
```

```python
#There's strong correlation
plt.scatter(vehicles['City MPG'], vehicles['Combined MPG'])
```

```python
# Its almost a perfect correlation
vehicles['City MPG'].corr(vehicles['Combined MPG'])
```

```python
# It's always one, that is correlation btw itself
vehicles['City MPG'].corr(vehicles['City MPG'])
```

```python
# It will display a correlation of all the numerical variables and leave out the ones that are categorical
vehicles.corr()
```
