### Bernoulli Distribution ###

```python
# This describes variable as an experiment where  the outcome can b classified as either success or failure
#This can only have two outcomes, the successful one and non-success
# We need only a single parameter
# P is always success
# The random variable X equal 1 if he outcom is a sucess and a = if it is a failure
```

```python
p = 0.5
bern = scipy.stats.bernoulli(p)
```

```python
# the probability of failure, which is 1 - p
bern.pmf(0)
```

```python
bern.pmf(4)
```

```python
bern.cdf(1)
```

```python
numrange = range(-10, 10)
plt.plot(numrange, bern.pmf(numrange), 'o')
```

```python
plt.plot(numrange, bern.cdf(numrange))
```

```python
x = bern.rvs(size = 10)
plt.hist(x)
```

```python
x = bern.rvs(size = 1000)
plt.hist(x)
```

### Binomial Distribution ###

```python
### Condiser a similar setup to Bernoulli, but here you are conducting 'n' such independent experiments and 
# recording the results of every experiment.

from scipy.stats import binom
```

```python
p = 0.5
n = 2
binomial = scipy.stats.binom(n,p)
```

```python
binomial.pmf(2)
```

```python
plt.plot(numrange, binomial.pmf(numrange), 'o')
```

```python
# CDF
binomial.cdf(2)
```

```python
plt.plot(numrange, binomial.cdf(numrange))
```

```python
x = binomial.rvs(size = 100)
plt.hist(x)
```

```python
p = 0.5
n = 20
binomial = scipy.stats.binom(n,p)
x = binomial.rvs(size = 100)
plt.hist(x)
```

```python
p = 0.5
n = 20
binomial = scipy.stats.binom(n,p)
x = binomial.rvs(size = 100)
plt.hist(x, bins = 20)
```

```python
# counting number of heads that comes out, that is from 0, to 20, the highest is 10 heads, that is the mass
numrange = range(0, 20)
plt.plot(numrange, binomial.pmf(numrange), 'o')
```

```python
x = range(20)
plt.plot(x, binomial.pmf(x), 'o')
plt.vlines(x, 0, binomial.pmf(x))
```

### Normal Distribution ###

```python
from scipy.stats import norm
```

```python
# the first parameter is mean and the second is standard deviation

normal = norm(0, 1)
```

```python
numrange = np.linspace(-5, 5, 100)
plt.plot(numrange, normal.pdf(numrange))
```

```python
# 
plt.plot(numrange, normal.cdf(numrange))
```

```python
x = normal.rvs(size = 100000)
plt.hist(x, bins = 1000)
plt.show()
```

```python
help(sns.lmplot)
```

```python
# for lmplot, we need three variable
sns.lmplot('x', 'y', data = df)
```

```python
# No regression line here
sns.lmplot('x', 'y', data = df, fit_reg = False)
```
