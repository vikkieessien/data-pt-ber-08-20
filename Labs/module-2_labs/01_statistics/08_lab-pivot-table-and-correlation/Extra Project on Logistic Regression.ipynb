#### Assess the performances of messages 0, 1 and 2. Which one should we use if we send notifications to all users? 

```python
Dummy_1 = pd.get_dummies(df_new['message_x'],drop_first=True)
Dummy_2 = pd.get_dummies(df_new['language_x'],drop_first = True)
Dummy_3 = pd.get_dummies(df_new['browser_x'],drop_first = True)
```

```python
# Merging the three dummies created for 'message', 'language_x', 'brower_y' 
    
f_final = pd.concat([df_new, Dummy_1, Dummy_2, Dummy_3], axis=1, join='outer')
f_final
```

##### Performances of messages 0, 1 and 2. Which one should we use if we send notifications to all users? 

```python
Y = f_final['Yes/No']
X = f_final[[1, 2]]
X = sm.add_constant(X)
model = sm.Logit(Y, X)
results = model.fit()
results.summary()
```

##### Message 1 above has more performance rate, with a coefficient of 0.26, and it should be condisered while sending out notifications to users

#### Translation manager thinks that one or several of the translations are crappy and look like a 4-year old could have written that. Can you confirm that? 
##### Translations to be considered

```python
Y = f_final['Yes/No']
X = f_final[['de', 'en', 'fr', 'ru']]
X = sm.add_constant(X)
model = sm.Logit(Y,X)
results = model.fit()
results.summary()
```

###### From the translations above, Deutch has more performance rate with a coefficient of 1.04, followed by English with 1.02 coefficient.
