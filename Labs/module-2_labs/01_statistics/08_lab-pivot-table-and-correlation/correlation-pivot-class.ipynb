### Correlation

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vehicles = pd.read_csv('../../03_data-visualization/02_lab-matplotlib-seaborn/your-code/vehicles.csv')
vehicles.head()
```

### Scatterplots

```python
plt.scatter(vehicles['City MPG'], vehicles['Highway MPG'])
```

### Linear Correlation between two variables

```python
vehicles['City MPG'].corr(vehicles['Highway MPG'])
```

### Correlation Matrix

```python
vehicles.corr()
```

### Correlation Heatmap

```python
sns.heatmap(vehicles.corr())
```

```python
sns.heatmap(vehicles.corr(), vmin=-1, vmax=1, annot=True, cmap="YlGnBu")
```

#### Spearman's Correlation Coefficient (Rank Correlation)

```python
vehicles.corr(method = 'spearman')
```

```python
sns.heatmap(vehicles.corr(method = 'spearman'), vmin=-1, vmax=1, annot=True, cmap="YlGnBu")
```
