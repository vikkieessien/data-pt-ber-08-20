```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
vehicles = pd.read_csv('vehicles.csv')
```

```python
vehicles.head()
```

```python
# sorting values by 
```

```python
#ploting the graph against City MPG and Highway MPG
vehicles.sort_values(by = ['City MPG', 'Highway MPG'],inplace = True)
plt.plot(vehicles['City MPG'], vehicles['Highway MPG'])
plt.show()
```

```python
# we added a variable to the column names
vehicles.plot(x = 'City MPG', y = 'Highway MPG')
plt.title('Ploting City MPG against Highway MPG')
plt.xlabel('City MPG')
plt.ylabel('Highway MPG')
plt.show()
```

### plotting the average City MPG by year ###

```python
group = vehicles[['City MPG', 'Year']].groupby(['Year']).mean().reset_index()

print(group)
```

```python
group = vehicles[['City MPG', 'Year']].groupby(['Year']).mean().reset_index()

#plt.show()
group.plot(x ='Year', y = 'City MPG')
plt.show()
```

```python
# the difference btw the two graphs is that the first has x, y axis assigned while the latter does not have x,y axis assigned
plt.plot(group['Year'], group['City MPG'])
plt.show()
```

Multiple Line Plot: Average City MPG and average Highway MPH 

```python
group2 = vehicles[['City MPG', 'Highway MPG', 'Year']].groupby(['Year']).mean().reset_index()
group2.plot(x = 'Year', y = ['City MPG', 'Highway MPG'])
plt.show()
```

```python
group2.plot.bar(x = 'Year', y = ['City MPG'])
plt.show()
```

```python
# adding more than one column
group2.plot.bar(x = 'Year', y = ['City MPG', 'Highway MPG'])
plt.show()
```
