### Dataframe Aggregation using groupby ###

```python
df.mean()
```

```python
df.groupby(['Fuel Type']).mean()
```

```python
df.groupby(['Fuel Type']).max()
```

```python
df.groupby(['Fuel Type']).min()
```

```python
df.groupby(['Fuel Type'])['Cylinders','City MPG'].agg(['mean', 'sum', 'std', 'max'])
```

```python
df.groupby(['Fuel Type'])['Cylinders', 'City MPG'].mean().reset_index()
```
